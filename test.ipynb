```python
import random
from typing import List
import math
```

```python
class Neuron:

    def __init__(self, input_size: int, bias: float=0.5) -> None:

        self.weights = [random.uniform(-1, 1) for _ in range(input_size)]
        self.bias = bias
    
    def forward(self, inputs: List[float]) -> float:

        if len(inputs) != len(self.weights):
            raise ValueError(f"Expected {len(self.weights)} input, got {len(inputs)} instead!")
        
        out = sum([i * w for i, w in zip(inputs, self.weights)]) + self.bias
        return out
```

```python
n = Neuron(3)
n.forward([0.5, 0.6, 0.6])
```

```python
class Layer:

    def __init__(self, input_size: int, n_neurons: int) -> None:
        
        self.neurons = [Neuron(input_size) for _ in range(n_neurons)]
    
    def forward(self, inputs: List[float]) -> List[float]:

        out = [n.forward(inputs) for n in self.neurons]
        return out
```

```python
l = Layer(2, 3)
l.forward([0.1, 0.9])
```

```python
class Activation:

    def __init__(self) -> None:
        pass

    def sigmoid(self, x: float) -> float:

        out = 1 / (1 + math.e ** -x)
        return out

    def forward(self, inputs: List[float]) -> List[float]:

        out = [self.sigmoid(x) for x in inputs]
        return out # makes layers output in 0 to 1
```

```python
l = Layer(2, 3)

a = Activation()
a.forward(l.forward([0.2, 0.9]))
```

```python
class Network:

    def __init__(self, layers: List[Layer]) -> None:
        
        self.layers = layers
    
    def forward(self, inputs):
        current = inputs
        for layer in self.layers:
            current = Activation().forward(layer.forward(current))       
        return current
```

```python
net = Network(layers=[
    Layer(n_neurons=4, input_size=2),
    Layer(n_neurons=3, input_size=4),
    Layer(n_neurons=1, input_size=3),
])

answer = net.forward([0.5, 0.3])
answer
```
